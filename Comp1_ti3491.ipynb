```python
param_dist = {
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2], 
    'lambda': [1, 1.5, 2],  
    'alpha': [0, 0.5, 1]  
}
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = df1.drop(columns=['NObeyesdad'])  
y = df1['NObeyesdad']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_valid, y_train, y_valid = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')


random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=25,
    scoring='accuracy',
    n_jobs=-1,
    cv=3,
    verbose=3,
    random_state=42
)

random_search.fit(X_train, y_train)

```

```python

from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score


eval_set = [(X_train, y_train), (X_valid, y_valid)]

model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', **random_search.best_params_)

model.fit(X_train, y_train, early_stopping_rounds=10, eval_set=eval_set, verbose=True)

train_predictions = model.predict(X_train)
train_accuracy = accuracy_score(y_train, train_predictions)
print("Training Accuracy: %.2f%%" % (train_accuracy * 100.0))

valid_predictions = model.predict(X_valid)
valid_accuracy = accuracy_score(y_valid, valid_predictions)
print("Validation Accuracy: %.2f%%" % (valid_accuracy * 100.0))

```

```python
import pandas as pd


x_test = test_df.drop('id', axis=1) 


x_test = pd.get_dummies(x_test, drop_first=True)


missing_cols = set(X_train.columns) - set(x_test.columns)  
for c in missing_cols:
    x_test[c] = 0
x_test = x_test[X_train.columns] 

predictions = model.predict(x_test)

test_df['NObeyesdad'] = label_encoder.inverse_transform(predictions)  

print(test_df.head())

```
